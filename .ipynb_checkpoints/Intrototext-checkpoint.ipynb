```python
import os

# Set the folder path where your .txt files are stored.
folder_path = "colonized"  # make sure the folder name is exactly as it is

# Create a dictionary to hold the texts.
texts = {}

# Loop over each file in the folder.
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        # Open and read the file; use errors='ignore' to skip decoding errors.
        with open(os.path.join(folder_path, filename), 'r', encoding='utf-8', errors='ignore') as f:
            texts[filename] = f.read()

# List the file names that have been loaded.
print("Loaded files:")
print(list(texts.keys()))
```

```python
# Manually classify the files based on the leader's type.
# (Adjust the filenames as they appear in your texts dictionary.)

political_files = [
    "BenazirBhutto.txt", "DesmondTutu.txt", "EllenJohnsonSirleaf.txt",
    "FidelCastro.txt", "ImranKhan.txt", "JawaharlalNehru.txt",
    "JohnDramaniMahama.txt", "KwameNkrumah.txt", "FidelCastro_MyLife.txt", 
    "NoFutureWithoutForgiveness.txt", "MyExperimentsWithTruth.txt"  # etc.
]

cultural_files = [
    "ChogyamTrungpa.txt", "FrantzFanon.txt", "JomoKenyatta.txt",
    "NgugiWaThiongo.txt", "AssataShakur.txt"  # etc.
]

# Check how many files are in each category.
print("Political Files:", political_files)
print("Cultural Files:", cultural_files)
```

```python
import nltk
from nltk import pos_tag, word_tokenize

# Download necessary NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def extract_verbs(text):
    # Tokenize the text into words.
    tokens = word_tokenize(text)
    # Tag the tokens with part-of-speech labels.
    tagged_tokens = pos_tag(tokens)
    # Extract words that are verbs (POS tags that start with 'VB').
    verbs = [word.lower() for word, tag in tagged_tokens if tag.startswith('VB')]
    return verbs

# Combine texts for each category
political_text = " ".join([texts[filename] for filename in political_files if filename in texts])
cultural_text  = " ".join([texts[filename] for filename in cultural_files if filename in texts])

# Extract verbs from each combined text.
poli_verbs = extract_verbs(political_text)
cult_verbs = extract_verbs(cultural_text)

# Check the results:
print("Number of verbs in political texts:", len(poli_verbs))
print("Number of verbs in cultural texts:", len(cult_verbs))
```

```python
import nltk
from nltk import FreqDist, word_tokenize

# Ensure NLTK resources are available
nltk.download('punkt')

# Define target words (all in lowercase for matching)
target_words = ['freedom', 'democracy', 'culture', 'leader']

# Loop over each text file
for filename, content in texts.items():
    # Convert text to lowercase and tokenize into words
    tokens = word_tokenize(content.lower())
    # Create a frequency distribution
    fdist = FreqDist(tokens)
    
    # Print the file name and frequency for each target word
    print(f"Frequencies in {filename}:")
    for word in target_words:
        print(f"  {word}: {fdist[word]}")
    print("-" * 40)
```

```python
import nltk
from nltk import ConditionalFreqDist, word_tokenize, pos_tag

# Define a function to extract verbs from text.
def extract_verbs(text):
    tokens = word_tokenize(text)
    tagged = pos_tag(tokens)
    verbs = [word.lower() for word, tag in tagged if tag.startswith('VB')]
    return verbs

# Create a list of pairs: (leader_type, verb)
pairs = []
# Political texts
for word in extract_verbs(poli_text):
    pairs.append(('political', word))
# Cultural texts
for word in extract_verbs(cult_text):
    pairs.append(('cultural', word))

# Build a Conditional Frequency Distribution.
cfd = ConditionalFreqDist(pairs)

# Display the most common verbs for each category.
print("Political leaders' top verbs:")
print(cfd['political'].most_common(10))
print("\nCultural leaders' top verbs:")
print(cfd['cultural'].most_common(10))

# Plot the CFD for a few selected verbs to see their usage.
cfd.plot(conditions=['political', 'cultural'], title="Verb Usage by Leader Type")
```

```python
def normalized_frequency(freq_dist, total_words, scale=1000):
    # Returns a dictionary with frequencies per `scale` words.
    return {word: (count / total_words) * scale for word, count in freq_dist.items()}

# Calculate total words for each group.
total_poli_words = len(word_tokenize(poli_text))
total_cult_words = len(word_tokenize(cult_text))

# Normalize the verb frequencies.
poli_norm = normalized_frequency(Counter(extract_verbs(poli_text)), total_poli_words)
cult_norm = normalized_frequency(Counter(extract_verbs(cult_text)), total_cult_words)

print("Normalized political verb frequencies (per 1000 words):")
print(sorted(poli_norm.items(), key=lambda x: x[1], reverse=True)[:10])

print("\nNormalized cultural verb frequencies (per 1000 words):")
print(sorted(cult_norm.items(), key=lambda x: x[1], reverse=True)[:10])
```

```python
import os

# Set the folder path exactly (make sure it matches your folder name)
folder_path = "colonized"

# Dictionary to hold file contents
texts = {}

# Loop over each file in the folder and load if it ends with .txt
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        with open(os.path.join(folder_path, filename), 'r', encoding='utf-8', errors='ignore') as f:
            texts[filename] = f.read()

# Print the filenames loaded to confirm
print("Loaded files:", list(texts.keys()))
```

```python
# Combine the contents of all files into one big string
combined_text = " ".join(texts[filename] for filename in texts)
print("Total length of combined text:", len(combined_text))
```

```python
import nltk
from nltk import word_tokenize, pos_tag
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
def extract_verbs(text):
    # Tokenize the text into words
    tokens = word_tokenize(text)
    # Tag each token with a part-of-speech label
    tagged_tokens = pos_tag(tokens)
    # Filter out tokens whose tag starts with 'VB' (verb forms)
    verbs = [word.lower() for word, tag in tagged_tokens if tag.startswith('VB')]
    return verbs

# Extract verbs from the combined text
all_verbs = extract_verbs(combined_text)
print("Total number of verbs extracted:", len(all_verbs))
```

```python
from collections import Counter

# Count the frequency of each verb
verb_freq = Counter(all_verbs)

# Display the top 10 most common verbs
top_verbs = verb_freq.most_common(10)
print("Top 10 most common verbs in the combined corpus:")
for verb, count in top_verbs:
    print(f"{verb}: {count}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dataset
data = {'Year': [2018, 2019, 2020, 2021],
        'Sales': [100, 150, 200, 250]}
df = pd.DataFrame(data)

# Plot the data
plt.plot(df['Year'], df['Sales'], marker='o')
plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# Create data
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [15, 25, 30, 35]

# Plot multiple lines
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='green', linestyle='--')

# Add labels, title, legend, and grid
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Comparison of Lines")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from textblob import TextBlob

# Sample story split into chapters or sections
sections = [
    "The knight prepared for battle with courage and determination.",
    "The dragon destroyed everything in its path, causing despair.",
    "After a fierce fight, the knight emerged victorious and brought hope."
]

# Analyze polarity for each section
polarities = [TextBlob(section).sentiment.polarity for section in sections]

# Plot sentiment
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(sections) + 1), polarities, marker='o', linestyle='-', color='blue')
plt.title("Sentiment Polarity Across Story Sections")
plt.xlabel("Section")
plt.ylabel("Sentiment Polarity")
plt.xticks(range(1, len(sections) + 1))  # Label x-axis with section numbers
plt.grid()
plt.show()
```

```python
# Sample story split into sentences
sentences = [
    "The knight was brave and ready for battle.",
    "The dragon burned the village to ashes.",
    "The villagers celebrated the knight's victory."
]

# Analyze sentiment for each sentence
for i, sentence in enumerate(sentences, 1):
    blob = TextBlob(sentence)
    print(f"Sentence {i}: {sentence}")
    print(f"  Polarity: {blob.sentiment.polarity}")
    print(f"  Subjectivity: {blob.sentiment.subjectivity}")
```
